```python
import pytesseract
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('text1.png')
b,g,r = cv2.split(image)
rgb_img = cv2.merge([r,g,b])
plt.imshow(rgb_img)
plt.title('Text1')
plt.show()
```

```python
custom_config = r'--oem 3 --psm 6'
print(pytesseract.image_to_string(image, config=custom_config))
```
