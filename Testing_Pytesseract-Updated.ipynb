```python
import pytesseract
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os

img =[]
for file_name in os.listdir(r"C:\Users\kunda\OneDrive\Documents\DataScience\Projects\OCR\Image"):
    if file_name.split(".")[-1].lower() in {"jpeg", "jpg", "png"}:
        img.append(file_name)      
img
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
def image_display(imag):
    image = cv2.imread(imag)
    b,g,r = cv2.split(image)
    rgb_img = cv2.merge([r,g,b])
    plt.imshow(rgb_img)
    plt.title(imag.split('.')[0])
    plt.show()

    print("------------------------------------------------------")
    custom_config = r'--oem 3 --psm 6'
    print(pytesseract.image_to_string(image, config=custom_config))

    print("******************************************************")
```

```python
for i,pic in enumerate(img):
    image_display(pic)
```
